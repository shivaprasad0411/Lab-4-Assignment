# **Question - 2**

```python
# importing the data from csv file
data2 = pd.read_csv("/content/MergerdAyush_Colleges_in_India_01-04-2021_0.csv")
data2.head()
```

```python
data2['State or Union Territory'] = data2['State or Union Territory'].str.strip()
```

```python
different_cat = data2.iloc[:,1].unique()
different_cat
```

```python
# new column which sums the government and non-government colleges
data2["Total colleges"] = data2["No. of Colleges - Govt"] + data2["No. of Colleges - Non-Govt"]
```

```python
# function to extract hex codes for color palette in sns
def extract_color_ids(palette):
  rgb_tuples = [tuple(int(255 * x) for x in color) for color in palette]
  hex_codes = [f"#{r:02x}{g:02x}{b:02x}" for r, g, b in rgb_tuples]
  return hex_codes
```

```python
# dark bipolar sequential color palette
image = Image.open("/content/bipolar_dark.png")
plt.imshow(image)
plt.axis("off")
plt.grid(False)

# we can use sns color palette to load dark bipolar sequential color palette
sns.color_palette("icefire")
plt.show()
```

```python
# defining color codes for different types of color palettes
palette_one = sns.dark_palette("#69d", reverse=True)
palette_continuous = sns.color_palette("dark:salmon_r")
palette_bipolar = sns.color_palette("icefire")

one_codes = extract_color_ids(palette_one)
continuous_codes = extract_color_ids(palette_continuous)
bipolar_codes = extract_color_ids(palette_bipolar)

one_codes
```

### Plots for different types of projections

```python
fig,axes = plt.subplots(3,1,figsize=(40,75))
display_types = ["Screen Display","Presentation","Poster Print"]

xy_fontsizes = [10,18,28]
title_fontsizes = [12,22,36]
legend_fontsizes = [8,14,20]

for j in range(3):
  for i in range(5):
    sampled_data = data2[data2["Type of System"] == different_cat[i]]
    sampled_data = sampled_data.sort_values(by="State or Union Territory")
    axes[j].bar(x=sampled_data["State or Union Territory"],height=sampled_data["Total colleges"],color=one_codes[i],label=different_cat[i],edgecolor=one_codes[i])

  axes[j].legend(fontsize = legend_fontsizes[j])
  axes[j].set_xlabel("State or Union Territory",fontsize=xy_fontsizes[j])
  axes[j].set_ylabel("Total colleges (Govt + Non-Govt)",fontsize=xy_fontsizes[j])
  axes[j].set_title("Distribution of National, State and District roads({})".format(display_types[j]),fontsize=title_fontsizes[j])
  axes[j].set_xticklabels(np.sort(data2["State or Union Territory"].unique()),rotation="vertical",fontsize=xy_fontsizes[j])

plt.show()
```

### Plots for different types of color palettes

```python
fig,axes = plt.subplots(3,1,figsize=(35,60))
types = ["Continuous color palette","One color sequential","Bipolar sequential"]
palettes = {
    0 : continuous_codes,
    1 : one_codes,
    2 : bipolar_codes
}

for j in range(3):
  for i in range(5):
    sampled_data = data2[data2["Type of System"] == different_cat[i]]
    sampled_data = sampled_data.sort_values(by="State or Union Territory")
    axes[j].bar(x=sampled_data["State or Union Territory"],height=sampled_data["Total colleges"],color=palettes[j][i],label=different_cat[i],edgecolor=palettes[j][i])

  axes[j].legend(fontsize = 17)
  axes[j].set_xlabel("State or Union Territory",fontsize=16)
  axes[j].set_ylabel("Total colleges (Govt + Non-Govt)",fontsize=16)
  axes[j].set_title("Distribution of National, State and District roads({})".format(types[j]),fontsize=20)
  axes[j].set_xticklabels(np.sort(data2["State or Union Territory"].unique()),rotation="vertical",fontsize=14)

plt.show()
```
